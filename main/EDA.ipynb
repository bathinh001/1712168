# Xem phân bố số lượng phim theo khoảng thời gian: mình sẽ chia thành 4 khoảng (< 90 phút, 90 - 120 phút, 120 -180 phút, > 180 phút)

# Phân tích về genre

```python
unique_genres = movie_df['genres'].unique()
individual_genres = []
for genre in unique_genres:
    individual_genres.append(genre.split(';'))

individual_genres = list(itertools.chain.from_iterable(individual_genres))
individual_genres = set(individual_genres)
```

```python
dict_genres = {}
for genre in individual_genres:
    current_genre = movie_df['genres'].str.contains(genre).fillna(False)
    dict_genres[genre] = movie_df[current_genre].shape[0]
```

# Top movie mọi thời đại

Nếu chỉ sử dụng average_rating thì sẽ có đôi chút không công bằng (ví dụ: một số phim lượt vote ít nhưng điểm cao, một số phim lượt vote nhiều điểm thấp). Do đó mình sẽ đặt ra metric (dựa trên weight rating của IMDB):

$ score = R(\frac{v}{v+m}) + C(\frac{m}{v+m})$

Trong đó:
- v số vote count của phim;
- m số vote nhỏ nhất để có thể đưa vào list top;
- R là average rating của phim;
- C là average rating của toàn bộ data

Từ đó mình sẽ lấy top dựa trên score

```python
C = movie_df['average_rating'].mean()
```

```python
# Mình sẽ dùng mức cutoff là 90%
m = movie_df['vote_counts'].quantile(0.9)
```

```python
filtered_movie_df = movie_df.copy().loc[movie_df['vote_counts'] >= m]
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_counts']
    R = x['average_rating']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
filtered_movie_df['score'] = filtered_movie_df.apply(weighted_rating, axis=1)
filtered_movie_df = filtered_movie_df.sort_values('score', ascending=False)
```
